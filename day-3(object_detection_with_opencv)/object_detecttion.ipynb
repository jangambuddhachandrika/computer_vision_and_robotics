```python
# Classical Object Detection using OpenCV (Non-DL)

## 1. Haar Cascade Face Detection

import cv2

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
img = cv2.imread('/home/jangcha/computer_vision_and_robotics/day-3(object_detection_with_opencv)/lena.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
import matplotlib.pyplot as plt
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Detected Face')
plt.axis('off')  # Optional: Hides axis ticks
plt.show()
```

```python
## 2. HOG People Detection

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

img = cv2.imread('/home/jangcha/computer_vision_and_robotics/test.jpg')
(rects, _) = hog.detectMultiScale(img, winStride=(4,4), padding=(8,8), scale=1.05)

for (x, y, w, h) in rects:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Detected img')
plt.axis('off')  # Optional: Hides axis ticks
plt.show()
```

```python
## 3. Contour-Based Object Detection

img = cv2.imread('/home/jangcha/computer_vision_and_robotics/pic3.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(img, contours, -1, (0, 0, 255), 2)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('contour img')
plt.axis('off')  # Optional: Hides axis ticks
plt.show()
```
